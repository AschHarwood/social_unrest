```python
nyt_apple_acled = pd.read_csv('https://nyc3.digitaloceanspaces.com/politicalnews/political_violence/features/nyt_cases_deaths_acled_protests_apple_mob_5_26_21.csv?X-Amz-Algorithm=AWS4-HMAC-SHA256&X-Amz-Credential=KSJCKUT5V42U4OBV57JG%2F20210526%2Fnyc3%2Fs3%2Faws4_request&X-Amz-Date=20210526T204052Z&X-Amz-Expires=259200&X-Amz-SignedHeaders=host&X-Amz-Signature=f2e5545462015ad7c412fa3e7fd2240f4a540f302d181fa39ac510b9943e2fb7')
```

```python
nyt_apple_acled.set_index(['date', 'fips', 'sub-region', 'region'], inplace=True)
```

```python
nyt_apple_acled.reset_index(inplace=True)
```

```python

mean = nyt_apple_acled.groupby(['date', 'fips', 'sub-region', 'region']).mean()
```

```python
mean.to_csv('nyt_apple_acled_transport_type_removed_5_27_21.csv')
```

```python
mean.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
mean.reset_index(inplace=True)
```

```python
mean
```

```python
def add_zero(x):
  if len(x) == 4:
    fips = '0' + x
    return fips
  else:
    return x
```

```python
mean.fips = mean.fips.astype('str')
mean.fips = mean.fips.apply(lambda x: add_zero(x))
```
